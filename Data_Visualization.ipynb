```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the Penguins dataset
df = sns.load_dataset('penguins')

# Display the first 10 rows
print(df.head(10))
```

```python
# Drop missing values for simplicity
df.dropna(inplace=True)
```

```python
# Histogram for 'bill_length_mm' (beak)
plt.figure(figsize=(6, 4))
sns.histplot(df['bill_length_mm'], bins=20, kde=True, color='skyblue')
plt.title("Histogram of Bill Length")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Boxplot for 'flipper_length_mm' to detect outliers
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['flipper_length_mm'], color='lightgreen')
plt.title("Boxplot of Flipper Length")
plt.xlabel("Flipper Length (mm)")
plt.grid(True)
plt.show()
```

```python
# Correlation Heatmap (numeric features only)
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f") #.2-round the number to 2 decimal points     #f-floating point
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot colored by species
sns.pairplot(df, hue='species', corner=True, palette='Set2')
plt.suptitle("Pairplot of Penguins Dataset by Species", y=1.02)
plt.show()
```
